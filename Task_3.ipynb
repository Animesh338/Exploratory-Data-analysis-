# Creating The Loss data frame 

```python
df_loss=df[df['Profit']<0]
df_loss
```

```python
df_loss.shape
```

```python
df_loss.describe()
```

```python
Total_loss=df_loss['Profit'].sum()
print('The total loss in the superstore is',Total_loss)
```

```python
df_loss.groupby(by='Segment').sum()
```

## As we can clearly see that more loss has been in the consumer segment than the other two segments . Also we can see that more the discount higher is the loss ,so we have to reduce the discount to earn more profit . 

# Sorting the values 

```python
df_loss_category=df_loss.groupby(by='Sub-Category').sum()
df_loss_category.sort_values(by='Profit',ascending=True)
```

# As we can see that more loss is received  in the Binders category, machines category, and tables category when compared to other categories.

## So better to give discounts on which are getting less sold so that even they will start getting sold more.
